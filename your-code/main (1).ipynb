# Challenge 1 - Plot Probability Mass Function (PMF)

#### Background knowledge

[PMF](https://en.wikipedia.org/wiki/Probability_mass_function) shows the probability distribution of a **discrete random variable**. A [discrete random variable](https://en.wikipedia.org/wiki/Random_variable#Discrete_random_variable) has random numeric values that are not continuous. For example, the number of people in a household can only be integers but not floats. Therefore the number of people in a household is a discrete variable.

#### Question: We assume that the probability of clicking an Ad in a Youtube video is 0.15. We have a sample of 5 people who wathched the video and we want to plot the PMF for a binomial distribution.

*Hint: use binom from `scipy.stats.binom`. Your output should look like below:*

![binom 10](ch-2.png)

```python
n_people = 5
p_success = 0.15

# x-values: possible number of clicks
k_values = np.arange(0, n_people + 1)

# Binomial probabilities
probabilities = binom.pmf(k_values, n_people, p_success)

# Plot
plt.bar(k_values, probabilities, color='skyblue', edgecolor='black')
plt.title('Binomial Distribution: Probability of People Clicking the Add')
plt.xlabel('Number of Clicks on Add')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

#### Explain what you observe from the plot above

```python
# Your comment here
# Probabilities are higher for 0-1 click on an add by these 5 people. It gets less likely that the add is clicked with more potential clicks.
```

#### A website has an average of 300 visits per day. What is the probability of getting 320 visitors in a day?

*Hint: use `scipy.stats.poisson`*

```python
# Your code here
from scipy.stats import poisson

# Parameters for the accident problem
lambda_rate = 300 # number of visitors per day
k_target = 320
prob = poisson.pmf(k_target, lambda_rate)
print(f"Probability of {k_target} per day is: {prob.round(7)}")
```
